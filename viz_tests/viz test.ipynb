```python
econ = pd.merge(labor_v_gross_ts[['Average weekly earnings of all employees',
       '12-month percent change in average weekly earnings of all employees',
       'Real average weekly earnings of all employees, 1982-1984 dollars',
       '12-month percent change in real average weekly earnings of all employees']], cp, on='Month')
econ = pd.merge(econ, we, on='Month')
econ.head()
```

```python
count = df_count.reset_index()
count = count[['Month','action', 'Comedy', 'Romance', 'Sci-Fi', 'Mystery',
       'Documentary', 'Adventure', 'News', 'Biography', 'Crime', 'Thriller',
       'Family', 'Musical', 'Horror', 'Animation', 'History', 'Sport', 'Drama',
       'Western', 'War', 'Music', 'Fantasy', 'Short']]

count = pd.melt(count, id_vars=['Month'], value_vars=['action', 'Comedy', 'Romance', 'Sci-Fi', 'Mystery',
       'Documentary', 'Adventure', 'News', 'Biography', 'Crime', 'Thriller',
       'Family', 'Musical', 'Horror', 'Animation', 'History', 'Sport', 'Drama',
       'Western', 'War', 'Music', 'Fantasy', 'Short'],
        var_name='Good', value_name='Change')



g = sns.FacetGrid(count, col="Good", col_wrap=5, height=3)
g.map(sns.lineplot, "Month", "Change")
g.set(xticks=count.Month[0::12])
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)
```
