```python
import pandas as pd
import re
import pickle
import numpy as np
```

```python
genre = pd.read_pickle('final-clean-genre_totals_by_month.pkl')
cp = pd.read_pickle('final-clean-consumer_prices.pkl')
we = pd.read_pickle('final-clean-weekly_earnings.pkl')
print(genre.columns)
print(cp.columns)
print(we.columns)
```

```python
alldat = pd.merge(genre, cp, how='left', left_on='date', right_on='Month')
alldat = pd.merge(alldat, we, how='left', left_on='date', right_on='Month')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
alldat
```

```python
alldat.to_pickle("final-clean-merged_data.pkl")
alldat.to_csv("final-clean-merged_data.csv")
```
