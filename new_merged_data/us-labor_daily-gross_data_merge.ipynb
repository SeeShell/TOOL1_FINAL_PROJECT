```python
import pandas as pd
```

```python
labor = pd.read_pickle('../us_labor_statistics/final-clean-weekly_earnings.pkl')
labor.head()
```

```python
movie = pd.read_pickle('../movie_daily_gross_boxofficemojo/clean-boxofficemojo-daily_gross.pkl')
movie.head()
```

```python
movie.columns
```

```python
movie['Month'] = movie.apply(lambda row: str(row.Date)[0:8] + '01', axis=1)
```

```python
movie.sample(5)
```

```python
# test to get monthly sums for a particular genre
month_action_group = movie[movie.Action==True].groupby('Month', as_index=False)
action_monthly_gross = month_action_group.agg({'Daily': 'sum'})
action_monthly_count = month_action_group.agg({'Daily': 'count'})
# action_monthly_gross = action_monthly_gross.rename(columns={'sum':'action'})
# action_monthly_count = action_monthly_count.rename(columns={'count':'action_count'})
a_gross = pd.DataFrame({'Month': action_monthly_gross['Month'], 'Action': action_monthly_gross['Daily']})
a_count = pd.DataFrame({'Month': action_monthly_count['Month'], 'Action': action_monthly_count['Daily']})
a_gross.head()
```

```python
# Get each genre daily and add to new dataframe
genres = ['Comedy',
       'Romance', 'Sci-Fi', 'Adult', 'Mystery', 'Documentary', 'Adventure',
       'News', 'Biography', 'Crime', 'Thriller', 'Family', 'Musical',
       'Horror', 'Animation', 'History', 'Sport', 'Drama', 'Film-Noir',
       'Western', 'War', 'Music', 'Fantasy', 'Short']

genre_monthly_gross = a_gross
genre_monthly_count = a_count
genre_monthly_gross.head()

for g in genres:    
    month_group = movie[movie[g]==True].groupby('Month', as_index=False)
    monthly_gross = month_group.agg({'Daily': 'sum'})
    monthly_count = month_group.agg({'Daily': 'count'})
    temp_gross = pd.DataFrame({'Month': monthly_gross['Month'], 'Daily': monthly_gross['Daily']})
    temp_count = pd.DataFrame({'Month': monthly_count['Month'], 'Daily': monthly_count['Daily']})
    genre_monthly_gross[g] = temp_gross['Daily']
    genre_monthly_count[g] = temp_count['Daily']


genre_monthly_count.head()
```

```python
labor.dtypes
```

```python
genre_monthly_count['Month'] = pd.to_datetime(genre_monthly_count['Month'])
genre_monthly_gross['Month'] = pd.to_datetime(genre_monthly_gross['Month'])
genre_monthly_gross.dtypes
```

```python
labor_movie_gross = pd.merge(genre_monthly_gross, labor, how='left', on='Month')
labor_movie_count = pd.merge(genre_monthly_count, labor, how='left', on='Month')
labor_movie_gross.to_pickle('monthly_labor_genre-daily-gross.pkl')
labor_movie_count.to_pickle('monthly_labor_genre-daily-count.pkl')
labor_movie_gross.head()
```
