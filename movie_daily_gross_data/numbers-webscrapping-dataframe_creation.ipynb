```python
# Build date for URL and make requests for each day

month_and_days = {'01': [x for x in range(1,32)], '02': [x for x in range(1,29)], '03':[x for x in range(1,32)],
                  '04':[x for x in range(1,31)],'05':[x for x in range(1,32)], '06':[x for x in range(1,31)], 
                 '07':[x for x in range(1,32)], '08':[x for x in range(1,32)], '09':[x for x in range(1,31)],
                 '10': [x for x in range(1,32)], '11':[x for x in range(1,31)], '12':[x for x in range(1,32)]}

leap_years = [2008, 2016, 2020]
leap_year_month_days = {'01': [x for x in range(1,32)], '02': [x for x in range(1,30)], '03':[x for x in range(1,32)],
                  '04':[x for x in range(1,31)],'05':[x for x in range(1,32)], '06':[x for x in range(1,31)], 
                 '07':[x for x in range(1,32)], '08':[x for x in range(1,32)], '09':[x for x in range(1,31)],
                 '10': [x for x in range(1,32)], '11':[x for x in range(1,31)], '12':[x for x in range(1,32)]}

column_names = ['Movie', 'Distr', 'Genre', 'Source', 'Creative_Type', 'Prod_Method', 'MPAA', 'Gross', '%YD', 
               '%LW', 'Thr','Per_Thr', 'Prod_Budg', 'D']
all_data = []

for year in range(2015, 2021):
    # build url for each year/month/day
    url = 'https://www.the-numbers.com/box-office-chart/daily/'
    
    if year in leap_years:
        m_d = leap_year_month_days
    else:
        m_d = month_and_days
        
    y = str(year)
    
    for month in m_d:
        m = '/' + month
        for day in m_d[month]:
            if len(str(day)) == 1:
                d = '/' + '0' + str(day)
            else:
                d = '/' + str(day)
            
            date = y + m + d
            url_curr = url + date
            
            #get data
            page = requests.get(url_curr)
            soup = bsoup(page.text, 'lxml')
            
            data_table = soup.find_all("table", id="box_office_daily_table")
            
            try:
                all_trs = data_table[0].find_all("tr")
            except:
                print(date)
                print(url_curr)
            
            for i in range(1, len(all_trs)):

                all_tds = all_trs[i].find_all("td")
                entry = []

                for j in range(0, len(all_tds)):
                    entry.append(all_tds[j].text)       
                
                entry = entry[2:]
                entry.insert(0, date)

                all_data.append(entry)
    print(f'{year} processed...')
    
## Error below shows that not all categories were coming through
```

```python
# Use only columns showing up and make dataframe
column_names_short = column_names = ['Date','Movie', 'Distr', 'Gross', '%YD', 
               '%LW', 'Thr','Per_Thr', 'Prod_Budg', 'D']
df = pd.DataFrame(all_data, columns=column_names_short)
df.head(20)
```

```python
df.dtypes
```

```python
# Replace non-numeric values for numeric data columns
df['Gross'] = df['Gross'].str.replace('[^0-9]', '')
df['Prod_Budg'] = df['Prod_Budg'].str.replace('[^0-9]', '')
df['%YD'] = df['%YD'].str.replace('[^0-9]', '')
df['%LW'] = df['%LW'].str.replace('[^0-9]', '')
df['Thr'] = df['Thr'].str.replace('[^0-9]', '')
df['Per_Thr'] = df['Thr'].str.replace('[^0-9]', '')
df['D'] = df['D'].str.replace('[^0-9]', '')
df.head()
```

```python
# create year, month, and day columns
split_date = pd.DataFrame(columns = ['year', 'month', 'day'])
split_date[['year','month', 'day']] = df.Date.str.split('/', n= 3, expand = True)
split_date.sample(10)
```

```python
df[['year', 'month', 'day']] = split_date[['year','month', 'day']]
```

```python
df.sample(20)
```

```python
# Replace white space with NaN and do a sanity check
df = df.replace(r'^\s*$', np.nan, regex=True)
df.iloc[6227]
```

```python
# Update numeric data types
df['Gross'] = pd.to_numeric(df['Gross'])
df['%YD'] = pd.to_numeric(df['%YD'])
df['%LW'] = pd.to_numeric(df['%LW'])
df['Thr'] = pd.to_numeric(df['Thr'])
df['Per_Thr'] = pd.to_numeric(df['Per_Thr'])
df['Prod_Budg'] = pd.to_numeric(df['Prod_Budg'])
df['D'] = pd.to_numeric(df['D'])
df.dtypes
```
